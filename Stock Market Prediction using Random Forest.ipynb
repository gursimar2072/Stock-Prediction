## Data Manipulation:-

```python
df.info()
```

```python
df.columns
```

```python
df["Date"] = df["Date"].astype(str)
```

```python
df.info()
```

```python
dd = df["Date"].str.split("-" ,n = -1,expand = True)
df["Year"] = dd[0]
df["month"] = dd[1]
df["Day"] = dd[2]
```

```python
df["Year"] = df["Year"].astype(int)
df["month"] = df["month"].astype(int)
df["Day"] = df["Day"].astype(int)
```

```python
df.info()
```

```python
X = df[["Year","month","Day","Open"]].values
X
```

```python
y = df["Close"].values
y
```

## Training Set and Testing Set in the Model:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=50)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

## Predictions and Scores

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
regressor.score(X_train,y_train)
```

```python
y_pred = regressor.predict([[2017,2,21,254.25]])
y_pred
#Predicted Values
```

#### Decision Tree Regressor:

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state =0 )
dtr.fit(X, y)
dtr.score(X,y)
```

```python
y_pred = dtr.predict([[2017,2,21,254.25]])
y_pred
#Predicted Values
```

#### Random Forest Regressor:

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators =300, random_state = 0)
rfr.fit(X, y)
rfr.score(X,y)
```

```python
y_pred = rfr.predict([[2017,2,21,254.25]])
y_pred
#Predicted Values
```
